```python
from time import time
```

```python
def is_prime(n, current_primes):
    if n < 2:
        return False
    for i in current_primes:
        if n % i == 0:
            return False
    return True

def count_primes(num):
    primes = 0
    current_primes = []
    # loop through numbers 1 to 100 as candidate primes
    for i in range(2, num):
        if is_prime(i + 1, current_primes):
            primes += 1
    return primes
```

```python
from tqdm.auto import tqdm
times = []
for i in tqdm(range(1000)):
    i += 1
    start = time()
    count = count_primes(i)
    end = time()
    times.append(end - start)
```

```python
import matplotlib.pyplot as plt
plt.plot(times)
```
